```python
import re
```

```python
# подключение к диску
from google.colab import drive
drive.mount('/content/gdrive', force_remount=True)

# путь до файла
path = '/content/gdrive/My Drive/COLAB_Andemir/classworks/cities.txt'
```

```python
# читаем файл

with open(path, encoding='CP866') as f:
  text = f.read()
f.close()  
```

```python
city = re.findall('<TR><td>([^,]+)', text)

# список городов
city_low = [i.lower() for i in city]
```

```python
k = True
check = []
user1 = input('User 1, where are we flying? ')
check.append(user1.lower())

for _ in range(1):
  if user1.lower() not in city_low:
    print('Show visa, user 2 flies without you')
    break 
  while True:
    # обработка варианта второго игрока
    if k:
      user2 = input('User 2, where are we flying? ')
      if user2.lower() in check:
        print('We were there already, user 1 flies without you')
        break
      if user2.lower() not in city_low:
        print('Show visa, user 1 flies without you')
        break 
      if user1[-1].lower() == user2[0].lower():
        check.append(user2.lower()) 
        k = not k
      else:
        print('You mixed up the flight, user 1 flies without you')
        break     
  # обработка варианта первого игрока
    else:
      user1 = input('User 1, where are we flying? ')
      if user1.lower() in check:
        print('We were there already, user 2 flies without you')
        break
      if user1.lower() not in city_low:
        print('Show visa, user 2 flies without you')
        break 
      if user2[-1].lower() == user1[0].lower():
        check.append(user1.lower()) 
        k = not k
      else:
        print('You mixed up the flight, user 2 flies without you')
        break         
```
